```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Maternal Health Risk Data Set.csv")
```

```python
risk_level = df['RiskLevel']
```

```python
##high risk , mid risk, low risk
```

```python
dummies = pd.get_dummies(risk_level)
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
df.drop('RiskLevel', axis = 1, inplace = True)
```

```python
df.to_csv("One hot encoded.csv", index = False)
```

```python
## There are 3 possible values which are high risk, low risk and midrisk
df['high risk'] = df['high risk'].astype(int)
df['low risk'] = df['low risk'].astype(int)
df['mid risk'] = df['mid risk'].astype(int)
df
```

```python
num_classes = 3
num_features  = df.shape[1]-num_classes ## Since there are 3 possible output values
```

```python
W = np.random.randn(num_classes, num_features)   # (K, d)
b = np.random.randn(num_classes)      # (K,)

x = np.random.randn(num_features)      # (d,)

scores = b + W @ x
scores
```

```python
def softmax(logits):
    """
    Computes the softmax activation for a given array of logits.

    Args:
        logits (np.ndarray): A 1D or 2D NumPy array of raw scores (logits).

    Returns:
        np.ndarray: A NumPy array representing the probability distribution.
    """
    # Subtract the maximum value for numerical stability to prevent overflow
    # when computing the exponential of large numbers.
    exp_logits = np.exp(logits - np.max(logits, axis=-1, keepdims=True))
    
    # Divide by the sum of exponentials to normalize into a probability distribution.
    probabilities = exp_logits / np.sum(exp_logits, axis=-1, keepdims=True)
    
    return probabilities
```

```python
scores_prob = softmax(scores)
scores_prob
```
